Bài 57: Với tập tin dữ liệu CSV đã được chuẩn hóa như trên bài 3 và bài 4, viết chương trình chia trường Giới tính có 2 giá trị thành 2 trường: “Nam” và “Nữ”, nhận giá trị 0 hoặc 1. Chuyển đổi dữ liệu giá trị cho trường Giới tính: giá trị “Nam”  là: 1 và “Nữ” là: 0. Xóa các bản ghi trùng lặp, chỉ giữ lại 1 bản ghi. Lưu và hiển thị dữ liệu kết quả như hình sau:

```python
import pandas as pd
```

```python
hoten = ['Nguyễn Thị Lệ', 'Lê Văn Tài', 'Hồ Hữu Hoàn', 'Lý Thị Lựu', 'Trần Trọng Chí', 'Phè Văn Phởn']
ngaysinh = ['01/01/45', '21/10/76', '03/03/90', '30/04/01', '01/09/63', '22/09/04']
gioitinh = [0, 1, 1, 0, 1, 1]
td_CN = [1, 1, 0, 1, 1, 1]
td_ThS = [1, 2, 0, 1, 0, 1]
td_TS = [0, 1, 0, 0 , 0, 1]
sodt = ['0915030909', '0902524444', '0232321234', '0170239311', '0904031234', '0907234445']
Nam = [0, 1, 1, 0, 1, 1]
Nu = [1, 0, 0, 1, 0, 0]
```

```python
Ds = zip(hoten, ngaysinh, gioitinh, td_CN, td_ThS, td_TS, sodt, Nam, Nu)
```

```python
df = pd.DataFrame(data = Ds, columns=['hoten', 'ngaysinh', 'gioitinh', 'td_CN', 'td_ThS', 'td_TS', 'sodt', 'Nam', 'Nu'])
df.to_csv('bai57.csv', index=False, header=True)
```
