```python
my_dict = {
  "name": "John",
  "yearofbirth": 2000,
  "class": "12A1",
  "math": 9.5,
  "english": 10
}

for x in my_dict:
    print("key:", x, "-", "value:", my_dict[x])
```

```python
# loop through my_dict using keys()
print("keys in my_dict:", end=' ')
for x in my_dict.keys():
    print(x, end=' ')

print( "\n------------------------------------")
    
# loop through my_dict using values()
print("\nvalues in my_dict:", end=' ')
for x in my_dict.values():
    print(x, end=' ')

print( "\n------------------------------------")

# loop through my_dict using items()
print("\nkeys and values in my_dict:")
for x, y in my_dict.items():
    print("key:", x, "-", "value:", y)
```

```python
# Sao chép cách 1
print("my_dict:", my_dict)
my_dict_copy = my_dict.copy()
print("my_dict_copy:", my_dict_copy)
```

```python
# sao chép cách 2
print("my_dict:", my_dict)
my_dict_copy = dict(my_dict)
print("my_dict_copy:", my_dict_copy)
```

```python
dictCar = {
    "brand": "Honda",
    "model": "Honda Civic",
    "year": 1972
}
for x in dictCar:
    print(x, ": ", dictCar.get(x))
```

```python
dictCar = {
    "brand": "Honda",
    "model": "Honda Civic",
    "year": 1972
}
dictCar.popitem()
print(dictCar)
```

```python
dictCar = {
    "brand": "Honda",
    "model": "Honda Civic",
    "year": 1972
}
dictCar["color"] = "yellow"
print(dictCar)
```

```python
d = dict()
def create_dict():
    for key in range(1, 21):
        d[key] = key**2
    return d
create_dict()
for key, value in d.items():
    print("key:", key, "-", "value:", value)

print( "\n------------------------------------")

for key in d.keys():
    print("key:", key)
    
print( "\n------------------------------------")

for value in d.values():
    print("value:", value)
```

```python
[key for key in d.keys()]
```
